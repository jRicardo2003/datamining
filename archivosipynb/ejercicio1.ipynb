```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url ="https://raw.githubusercontent.com/jRicardo2003/datamining/refs/heads/main/archivos/Ejercicio%201.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df_numeric = df.select_dtypes(include=['number'])
```

```python
matriz = df_numeric.corr()
matriz
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(matriz, annot=True, cmap="coolwarm")
plt.title("Matriz de correlación")
plt.show()
```
